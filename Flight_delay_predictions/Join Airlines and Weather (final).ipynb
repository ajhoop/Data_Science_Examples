## Need  cross reference  for special-cases

Most stations will join based on the 'call' column to the DEST and ORIGIN in the airlines data

```
# File location and type
file_location = "/FileStore/tables/airport_code_iata_icao.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
airport_iata_icao = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

airport_iata_icao.createOrReplaceTempView("airport_iata_icao")
airport_iata_icao.createOrReplaceTempView("airport_iata_icao")
display(airport_iata_icao)
```

```
airport_iata_icao.createOrReplaceTempView("airport_iata_icao")
airport_iata_icao = airport_iata_icao.withColumnRenamed('City served','CITY')
airport_iata_icao = airport_iata_icao.withColumnRenamed('Airport name','AIRPORT_NAME')
airport_iata_icao.createOrReplaceTempView("airport_iata_icao")
```

```
display(spark.sql("select * from airport_iata_icao where CITY like 'Will%' "))
```

```
airport_xref = spark.sql( "select * from airport_iata_icao where CITY is not null")
airport_xref.createOrReplaceTempView("airport_xref")
```

```
stations.createOrReplaceTempView("stations")
display(spark.sql("select * from stations where (name like 'SAN FRAN%' and  state = 'CA' and call is not null)" ))
```

### Join to cross reference to get IATA code for stations in non-contigous states.

```
stations_xref = stations.join(airport_xref, stations.call == airport_xref.ICAO, how="left")
stations_xref.where("ICAO is not null").count()
```

```
stations_xref.createOrReplaceTempView("stations_xref")
stations_final = spark.sql("select  usaf || wban as STATION_ID, *, case when IATA is null then substr(CALL,2,4) else IATA  end as AIRPORT_CODE   from stations_xref ")
stations_final.createOrReplaceTempView("stations_final")
stations_final = spark.sql("select AIRPORT_CODE, STATION_ID, call, IATA, lat, lon, elev from stations_final where AIRPORT_CODE is not null and (country = 'US' or IATA is not null)  AND (IATA is not null or call not like 'P%')")
stations_final.createOrReplaceTempView("stations_final")
```

```
print('stations with airport_code', spark.sql("select count(*) from stations_final where AIRPORT_CODE is not null").first()[0])
```

# Join Stations to Weather

```
weather.createOrReplaceTempView("weather")
print("stations ", stations_final.count())
unique_stations = spark.sql("select STATION, count(*) as weather_count from weather group by STATION")
unique_stations.createOrReplaceTempView("unique_stations")

stations_with_weather_obs = stations_final.join(unique_stations, unique_stations.STATION == stations_final.STATION_ID, how='inner')
stations_with_weather_obs = stations_with_weather_obs.drop("STATION")
stations_with_weather_obs.cache()
stations_with_weather_obs.createOrReplaceTempView("stations_with_weather_obs")
print("stations with weather observations", stations_with_weather_obs.count())
```

```
stations_with_weather_obs.createOrReplaceTempView("stations_with_weather_obs")
distinct_dest = spark.sql("select DEST, DEST_CITY_NAME, count(*) as DEST_COUNT from airlines group by DEST, DEST_CITY_NAME order by DEST_CITY_NAME")
distinct_dest.createOrReplaceTempView("distinct_dest")
dest_to_station = distinct_dest.join(stations_with_weather_obs, distinct_dest.DEST == stations_with_weather_obs.AIRPORT_CODE, how='left')
dest_to_station.createOrReplaceTempView("dest_to_station")
display(spark.sql("select DEST, DEST_CITY_NAME, DEST_COUNT from dest_to_station where AIRPORT_CODE is null order by DEST, DEST_CITY_NAME"))
stations_with_weather_obs.createOrReplaceTempView("stations_with_weather_obs")
```

```
stations_with_weather_obs.createOrReplaceTempView("stations_with_weather_obs")
print("stations with weather observations", stations_with_weather_obs.count())

stations_w_weather_unique = stations_with_weather_obs.dropDuplicates(["CALL", "IATA"])

print("stations with weather observations, no dups", stations_w_weather_unique.count())
stations_w_weather_unique = stations_w_weather_unique.drop("IATA")
stations_w_weather_unique = stations_w_weather_unique.drop("call")
stations_w_weather_unique = stations_w_weather_unique.drop("weather_count")
stations_w_weather_unique.createOrReplaceTempView("stations_w_weather_unique")
```

```
display(spark.sql("select AIRPORT_CODE, count(*) as count from stations_w_weather_unique group by AIRPORT_CODE having count > 1 order by count desc "))
```

# Join Weather Stations to Airlines

#### Join airport weather stations to airlines for DEST and ORIGIN airports

```
print("count before joining to stations", airlines.count())
airlines_origin_station = airlines.join(stations_w_weather_unique, airlines.ORIGIN == stations_w_weather_unique.AIRPORT_CODE, how='left')

airlines_origin_station = airlines_origin_station.withColumnRenamed('lat','ORIGIN_LATITUDE')
airlines_origin_station = airlines_origin_station.withColumnRenamed('lon','ORIGIN_LONGITUDE')
airlines_origin_station = airlines_origin_station.withColumnRenamed('elev','ORIGIN_ELEV')
airlines_origin_station = airlines_origin_station.withColumnRenamed('STATION_ID','ORIGIN_STATION_ID')
airlines_origin_station = airlines_origin_station.drop('name')
```

```
print("count after joining to origin stations", airlines_origin_station.count())
```

```
airlines_origin_station_cloned = spark.createDataFrame(airlines_origin_station.rdd, airlines_origin_station.schema)

airlines_join_station = airlines_origin_station_cloned.join(stations_w_weather_unique, airlines_origin_station_cloned.DEST == stations_w_weather_unique.AIRPORT_CODE, how='left')

airlines_join_station = airlines_join_station.withColumnRenamed('lat','DEST_LATITUDE')
airlines_join_station = airlines_join_station.withColumnRenamed('lon','DEST_LONGITUDE')
airlines_join_station = airlines_join_station.withColumnRenamed('elev','DEST_ELEV')
airlines_join_station = airlines_join_station.withColumnRenamed('STATION_ID','DEST_STATION_ID')
airlines_join_station = airlines_join_station.drop('name')

print("count after joining to dest stations", airlines_join_station.count())
```

# Join Weather to Airlines

### Join Weather to Airport stations 
- This will reduce the weather dataset to only the stations we care about

```
print("weather count before joining to stations", weather.count())
weather_airports = weather.join(stations_w_weather_unique, weather.STATION == stations_w_weather_unique.STATION_ID, how='inner')
weather_airports.cache()
print("weather count before joining to stations", weather_airports.count())
```

#### Aggregation of weather will require transformations to parse out numeric representation

```
weather_airports.createOrReplaceTempView("weather_airports")
weather_daily = spark.sql("select airport_code as AIRPORT_CODE_W, year(DATE) as YEAR_W, "\
                  "month(DATE) as MONTH_W,  "\
                  "dayofmonth(DATE) as DAYOFMONTH_W, "\
                  "count(*) AS COUNT_W, "\
                  "AVG(case when substr(WND,1,3) == '999'    then null else int(substr(WND,1,3)) end) as WND_ANGLE_AVG, "\
                  "AVG(case when substr(WND,9,4) == '9999'   then null else int(substr(WND,9,4)) end) as WND_SPEED_AVG, "\
                  "AVG(case when substr(CIG,1,5) == '99999'  then null else int(substr(CIG,1,5)) end) as CIG_AVG, "\
                  "AVG(case when substr(VIS,1,6) == '999999' then null else int(substr(VIS,1,6)) end) as VIS_AVG, "\
                  "AVG(case when substr(TMP,1,5) == '+9999'  then null else int(substr(TMP,1,5)) end) as TMP_AVG, "\
                  "AVG(case when substr(DEW,1,5) == '+9999'  then null else int(substr(DEW,1,5)) end) as DEW_AVG, "\
                  "AVG(case when substr(SLP,1,5) == '99999'  then null else int(substr(SLP,1,5)) end) as SLP_AVG, "\
                  "AVG(case when AA1 is null                 then null else (int(substr(AA1,4,4)) / int(substr(AA1,1,2))) end) as PRECIP_RATE_AVG,  "\
                  "AVG(case when AJ1 != '' and substr(AJ1,10,6) != '999999'  then null else int(substr(AJ1,1,4)) end) as SNOW_DEPTH_AVG,  "\
                  "AVG(case when AJ1 != '' and substr(AJ1,10,6) != '999999'  then null else int(substr(AJ1,10,6)) end) as SNOW_DEPTH_WATER_AVG "\
                  "from weather_airports "\
                  "group by AIRPORT_CODE_W, YEAR_W, MONTH_W, DAYOFMONTH_W ")
```

```
weather_daily.createOrReplaceTempView("weather_daily")
display(weather_daily)
```

```
weather_count = weather_daily.count()
print('WEATHER DAILY', f'{weather_count:,}', '\n')
```

```
weather_daily_sample = weather_daily.sample(False, 0.001)
display(weather_daily_sample.describe())
```

```
# Drop SNOW_DEPTH_WATER_AVG
weather_daily_sample = weather_daily_sample.drop("SNOW_DEPTH_WATER_AVG")
```

### Join weather daily averages to airlines

```
airlines_stations_weather = airlines_join_station.join(weather_daily, 
                                        (weather_daily.AIRPORT_CODE_W == airlines_join_station.ORIGIN) & \
                                        (weather_daily.YEAR_W         == airlines_join_station.YEAR) & \
                                        (weather_daily.MONTH_W        == airlines_join_station.MONTH) & \
                                        (weather_daily.DAYOFMONTH_W   == airlines_join_station.DAY_OF_MONTH), \
                                        how='left')
airlines_stations_weather = airlines_stations_weather.withColumnRenamed('WND_ANGLE_AVG','ORIGIN_WND_ANGLE_AVG')
airlines_stations_weather = airlines_stations_weather.withColumnRenamed('WND_SPEED_AVG','ORIGIN_WND_SPEED_AVG')
airlines_stations_weather = airlines_stations_weather.withColumnRenamed('CIG_AVG','ORIGIN_CIG_AVG')
airlines_stations_weather = airlines_stations_weather.withColumnRenamed('TMP_AVG','ORIGIN_TMP_AVG')
airlines_stations_weather = airlines_stations_weather.withColumnRenamed('DEW_AVG','ORIGIN_DEW_AVG')
airlines_stations_weather = airlines_stations_weather.withColumnRenamed('SLP_AVG','ORIGIN_SLP_AVG')
airlines_stations_weather = airlines_stations_weather.withColumnRenamed('VIS_AVG','ORIGIN_VIS_AVG')
airlines_stations_weather = airlines_stations_weather.withColumnRenamed('PRECIP_RATE_AVG','ORIGIN_PRECIP_RATE_AVG')

airlines_stations_weather = airlines_stations_weather.withColumnRenamed('AA2_AVG','ORIGIN_AA2_AVG')
airlines_stations_weather = airlines_stations_weather.withColumnRenamed('SNOW_DEPTH_AVG','ORIGIN_SNOW_DEPTH_AVG')



airlines_stations_weather = airlines_stations_weather.drop('AIRPORT_CODE_W')
airlines_stations_weather = airlines_stations_weather.drop('YEAR_W')
airlines_stations_weather = airlines_stations_weather.drop('MONTH_W')
airlines_stations_weather = airlines_stations_weather.drop('DAYOFMONTH_W')
airlines_stations_weather = airlines_stations_weather.drop('COUNT_W')
airlines_stations_weather.count()
```

```
airlines_stations_weather_cloned = spark.createDataFrame(airlines_stations_weather.rdd, airlines_stations_weather.schema)
airlines_stations_weather1 = airlines_stations_weather_cloned.join(weather_daily, 
                                        (weather_daily.AIRPORT_CODE_W == airlines_stations_weather_cloned.DEST) & \
                                        (weather_daily.YEAR_W         == airlines_stations_weather_cloned.YEAR) & \
                                        (weather_daily.MONTH_W        == airlines_stations_weather_cloned.MONTH) & \
                                        (weather_daily.DAYOFMONTH_W   == airlines_stations_weather_cloned.DAY_OF_MONTH), \
                                        how='left')

airlines_stations_weather1 = airlines_stations_weather1.withColumnRenamed('WND_ANGLE_AVG','DEST_WND_ANGLE_AVG')
airlines_stations_weather1 = airlines_stations_weather1.withColumnRenamed('WND_SPEED_AVG','DEST_WND_SPEED_AVG')
airlines_stations_weather1 = airlines_stations_weather1.withColumnRenamed('CIG_AVG','DEST_CIG_AVG')
airlines_stations_weather1 = airlines_stations_weather1.withColumnRenamed('TMP_AVG','DEST_TMP_AVG')
airlines_stations_weather1 = airlines_stations_weather1.withColumnRenamed('DEW_AVG','DEST_DEW_AVG')
airlines_stations_weather1 = airlines_stations_weather1.withColumnRenamed('SLP_AVG','DEST_SLP_AVG')
airlines_stations_weather1 = airlines_stations_weather1.withColumnRenamed('VIS_AVG','DEST_VIS_AVG')
airlines_stations_weather1 = airlines_stations_weather1.withColumnRenamed('PRECIP_RATE_AVG','DEST_PRECIP_RATE_AVG')

airlines_stations_weather1 = airlines_stations_weather1.withColumnRenamed('SNOW_DEPTH_AVG','DEST_SNOW_DEPTH_AVG')



airlines_stations_weather1 = airlines_stations_weather1.drop('AIRPORT_CODE_W')
airlines_stations_weather1 = airlines_stations_weather1.drop('YEAR_W')
airlines_stations_weather1 = airlines_stations_weather1.drop('MONTH_W')
airlines_stations_weather1 = airlines_stations_weather1.drop('DAYOFMONTH_W')
airlines_stations_weather1 = airlines_stations_weather1.drop('COUNT_W')
airlines_stations_weather1 = airlines_stations_weather1.drop('AIRPORT_CODE')
airlines_stations_weather1.count()
```
